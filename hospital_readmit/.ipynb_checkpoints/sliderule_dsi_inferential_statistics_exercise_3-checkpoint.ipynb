# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

Hypothesis test:

To quantify correlation between readmissions and discharges, we can calculate the Pearson correlation coefficient r. The null hypothesis is that there is no correlation (r=0), and the alternative hypothesis that the actual correlation is non-zero. The significance of the correlation can be then judged by the p-value for the coefficient.

```python
# Scipy.stats.pearsonr() is supposed to return the pearson coefficient and the corresponding p-value
from scipy import stats
pearson_r, p_value = stats.pearsonr(x, y)
n = len(x)
print('Pearson r:', pearson_r, '\nP-value:', p_value, '\nDegrees of freedom:', n-2)
```

```python
# Verify that the small p_value makes sense by calculating it more explicitely
t_statistic = pearson_r/(np.sqrt(1 - pearson_r**2))*np.sqrt(n - 2) # t statistic for r
p_value = stats.t.sf(np.abs(t_statistic), n-2)*2 # 2-tailed
print('T_statistic:', t_statistic, "\nP-value:", p_value)
```

The value of the correlation coefficient is small (r = -0.093), but thanks to the large number of samples (n = 11500),  it is statistically significant with negligible p-value (p = 1.5e-23). Based on the significance level $\alpha = 0.01$, the null hypothesis should be rejected. There appears to be small, but statistically significant correlation.

The above frequentist test assumes bivariate normal distribution of the sample. Bootstrap test can break the correlation while retaining some characteristics of the marginal distributions.

```python
# define bootstaraping function for generating samples of uncorrlated pairs from a pair of samples
def bootstrap_correlation(x_in, y_in, bs_number=1):
    bs_corr = []
    for _ in range(bs_number):
        y_perm = np.random.permutation(y_in)
        bs_corr.append(np.corrcoef(x_in, y_perm)[0,1])
                          
    return np.array(bs_corr)
```

```python
bs_pearson_r = bootstrap_correlation(x, y, 10000)
bs_pvalue = np.sum(np.abs(bs_pearson_r) >= np.abs(pearson_r))/len(bs_pearson_r)
print('Bootstrap p-value:', bs_pvalue)
```

Again, the bootstrap test confirms that we should reject the null hypothesis. There is a statistically significant (negative) correlation between the readmission rates and the discharges (hospital size).

While the statistical significance of the correlation can be established, it is rather small in absolute value, and the relation between the quantities is not linear. 

Test correlation for larger hospitals only, to see if the correlation persists in this region.

```python
large_hospital = clean_hospital_read_df.loc[clean_hospital_read_df['Number of Discharges'] > 1000]
excess = large_hospital['Excess Readmission Ratio'].values
discharge = large_hospital['Number of Discharges'].values
pearson_r, p_value = stats.pearsonr(discharge, excess)
print('Large hospital (dicharges > 1000) correlation between discharges and readmissions')
print('Number of samples', excess.shape[0])
print('Pearson r:', pearson_r)
bs_pearson_r = bootstrap_correlation(discharge, excess, 10000)
bs_pvalue = np.sum(np.abs(bs_pearson_r) >= np.abs(pearson_r))/len(bs_pearson_r)
print('Bootstrap p-value:', bs_pvalue)
```

The correlation coefficient obtained only for large hospitals is smaller in absolute value, and the calculated p-value would not pass the hypothesis test at the desired level of $\alpha=0.01$ (or even 0.05). There does not seem to be statistically significant correlation once the hospital size reaches a certain level. One of the reasons may be the smaller number of large hospitals, which does not allow us to make such inference for the small magnitude of the coefficient.
